```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
data_path = '../13. Data from R 50000/'
X_train = pd.read_parquet(data_path + 'X_train_2024.parquet', engine='pyarrow')
X_test = pd.read_parquet(data_path + 'X_test_2024.parquet', engine='pyarrow')
seed_num = 2024

# Load y_train and y_test
y_train = pd.read_parquet(data_path + 'y_train_2024.parquet').iloc[:, 0]
y_test = pd.read_parquet(data_path + 'y_test_2024.parquet').iloc[:, 0]

# Function to perform KMeans clustering and segment data
def clustered_segmentation(X_train, X_test, age_column, n_clusters):
    kmeans = KMeans(n_clusters=n_clusters, random_state=seed_num, n_init=30)
    X_train['age_cluster'] = kmeans.fit_predict(X_train[[age_column]])
    X_test['age_cluster'] = kmeans.predict(X_test[[age_column]])
    
    train_clusters = {i: X_train[X_train['age_cluster'] == i] for i in range(n_clusters)}
    test_clusters = {i: X_test[X_test['age_cluster'] == i] for i in range(n_clusters)}
    
    return train_clusters, test_clusters, kmeans

# Evaluate model on each cluster
def train_evaluate_cluster(cluster_id, train_data, test_data, y_train, y_test, all_y_test, all_y_pred):
    if len(train_data) < 2 or len(test_data) < 2:
        print(f"Cluster {cluster_id} too small to evaluate, skipping this cluster.")
        return

    model = XGBRegressor(random_state=2024)
    model.fit(train_data.drop('age_cluster', axis=1), y_train)
    y_pred = model.predict(test_data.drop('age_cluster', axis=1))
    
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
    rmspe = np.sqrt(np.mean(np.square((y_test - y_pred) / y_test))) * 100
    r2 = r2_score(y_test, y_pred)

    print(f'Cluster {cluster_id} - Mean Squared Error: {mse}')
    print(f'Cluster {cluster_id} - Mean Absolute Error: {mae}')
    print(f'Cluster {cluster_id} - Root Mean Squared Error: {rmse}')
    print(f'Cluster {cluster_id} - Mean Absolute Percentage Error: {mape:.2f}%')
    print(f'Cluster {cluster_id} - Root Mean Squared Percentage Error: {rmspe:.2f}%')
    print(f'Cluster {cluster_id} - R-squared: {r2:.2f}')

    relative_errors = (y_pred - y_test) / y_test
    prediction_errors[cluster_id] = relative_errors
    
    within_10_percent = calculate_within_tolerance(y_test, y_pred, 0.10)
    within_50_percent = calculate_within_tolerance(y_test, y_pred, 0.50)
    within_100_percent = calculate_within_tolerance(y_test, y_pred, 1.00)

    print(f'Cluster {cluster_id} - Percentage within 10%: {within_10_percent:.2f}%')
    print(f'Cluster {cluster_id} - Percentage within 50%: {within_50_percent:.2f}%')
    print(f'Cluster {cluster_id} - Percentage within 100%: {within_100_percent:.2f}%')

    # Store results for full test set evaluation
    all_y_test.extend(y_test.tolist())
    all_y_pred.extend(y_pred.tolist())

# Function to calculate percentage within tolerance
def calculate_within_tolerance(y_true, y_pred, tolerance):
    within_tolerance = abs(y_true - y_pred) / y_true <= tolerance
    percentage_within_tolerance = within_tolerance.mean() * 100
    return percentage_within_tolerance

# Loop over different numbers of clusters
for n_clusters in range(2, 21):
    print(f'\n--- Evaluating {n_clusters} Clusters ---')

    # Perform clustering
    train_clusters, test_clusters, kmeans = clustered_segmentation(X_train.copy(), X_test.copy(), 'DRIVER1_AGE', n_clusters)
    
    # Initialize storage for metrics and segment sizes
    prediction_errors = {}
    all_y_test = []
    all_y_pred = []

    # Evaluate each cluster
    for cluster_id in train_clusters.keys():
        train_evaluate_cluster(
            cluster_id,
            train_clusters[cluster_id],
            test_clusters[cluster_id],
            y_train.loc[train_clusters[cluster_id].index],
            y_test.loc[test_clusters[cluster_id].index],
            all_y_test,
            all_y_pred
        )

    # Convert lists to numpy arrays for evaluation
    all_y_test = np.array(all_y_test)
    all_y_pred = np.array(all_y_pred)

    # Calculate overall metrics on the full test set
    if len(all_y_test) > 0:
        overall_mse = mean_squared_error(all_y_test, all_y_pred)
        overall_mae = mean_absolute_error(all_y_test, all_y_pred)
        overall_rmse = np.sqrt(overall_mse)
        overall_mape = np.mean(np.abs((all_y_test - all_y_pred) / all_y_test)) * 100
        overall_rmspe = np.sqrt(np.mean(np.square((all_y_test - all_y_pred) / all_y_test))) * 100
        overall_r2 = r2_score(all_y_test, all_y_pred)
        overall_within_10_percent = calculate_within_tolerance(all_y_test, all_y_pred, 0.10)
        overall_within_50_percent = calculate_within_tolerance(all_y_test, all_y_pred, 0.50)
        overall_within_100_percent = calculate_within_tolerance(all_y_test, all_y_pred, 1.00)

        print(f'\nOverall Metrics for {n_clusters} Clusters:')
        print(f'Mean Squared Error: {overall_mse}')
        print(f'Mean Absolute Error: {overall_mae}')
        print(f'Root Mean Squared Error: {overall_rmse}')
        print(f'Mean Absolute Percentage Error: {overall_mape:.2f}%')
        print(f'Root Mean Squared Percentage Error: {overall_rmspe:.2f}%')
        print(f'R-squared: {overall_r2:.2f}')
        print(f'Percentage within 10%: {overall_within_10_percent:.2f}%')
        print(f'Percentage within 50%: {overall_within_50_percent:.2f}%')
        print(f'Percentage within 100%: {overall_within_100_percent:.2f}%')
    else:
        print("No valid clusters found for overall evaluation.")

    # Plotting the density curve of relative prediction errors for all clusters
    plt.figure(figsize=(12, 8))
    for cluster_id, errors in prediction_errors.items():
        sns.kdeplot(errors, label=f'Cluster {cluster_id} (n={len(errors)})')
    plt.title(f'Density Plot of Relative Prediction Errors for {n_clusters} Clusters')
    plt.xlabel('Relative Prediction Error (Error / True Value)')
    plt.ylabel('Density')
    plt.legend()
    plt.show()
```

```python
# Function to perform KMeans clustering and add cluster as a feature
def add_cluster_feature(X_train, X_test, age_column, n_clusters):
    kmeans = KMeans(n_clusters=n_clusters, random_state=seed_num, n_init=30)
    X_train[f'age_cluster_{n_clusters}'] = kmeans.fit_predict(X_train[[age_column]])
    X_test[f'age_cluster_{n_clusters}'] = kmeans.predict(X_test[[age_column]])
    return X_train, X_test

# Function to calculate percentage within tolerance
def calculate_within_tolerance(y_true, y_pred, tolerance):
    within_tolerance = abs(y_true - y_pred) / y_true <= tolerance
    percentage_within_tolerance = within_tolerance.mean() * 100
    return percentage_within_tolerance

# Initialize a plot for combined density curves
plt.figure(figsize=(12, 8))

# Loop over different numbers of clusters
for n_clusters in range(2, 11):
    print(f'\n--- Evaluating with {n_clusters} Clusters as a Feature ---')

    # Add the cluster feature to the data
    X_train_with_cluster, X_test_with_cluster = add_cluster_feature(X_train.copy(), X_test.copy(), 'DRIVER1_AGE', n_clusters)
    
    # Train the model using the entire dataset with the cluster feature
    model = XGBRegressor(random_state=seed_num)
    model.fit(X_train_with_cluster, y_train)
    y_pred = model.predict(X_test_with_cluster)
    
    # Calculate metrics
    mse = mean_squared_error(y_test, y_pred)
    within_10_percent = calculate_within_tolerance(y_test, y_pred, 0.10)
    within_50_percent = calculate_within_tolerance(y_test, y_pred, 0.50)
    within_100_percent = calculate_within_tolerance(y_test, y_pred, 1.00)

    print(f'Mean Squared Error: {mse}')
    print(f'Percentage within 10%: {within_10_percent:.2f}%')
    print(f'Percentage within 50%: {within_50_percent:.2f}%')
    print(f'Percentage within 100%: {within_100_percent:.2f}%')

    # Calculate relative errors for plotting
    relative_errors = (y_pred - y_test) / y_test

    # Add the density plot for this clustering number to the combined plot
    sns.kdeplot(relative_errors, label=f'Dataset c{n_clusters}')

# Finalize and show the combined density plot
plt.title('Combined Density Plot of Relative Prediction Errors for Different Cluster Numbers')
plt.xlabel('Relative Prediction Error (Error / True Value)')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
# Load the datasets
data_path = '../13. Data from R 50000/'
X_train = pd.read_parquet(data_path + 'X_train_2024.parquet', engine='pyarrow')
X_test = pd.read_parquet(data_path + 'X_test_2024.parquet', engine='pyarrow')
seed_num = 2024

# Load y_train and y_test
y_train = pd.read_parquet(data_path + 'y_train_2024.parquet').iloc[:, 0]
y_test = pd.read_parquet(data_path + 'y_test_2024.parquet').iloc[:, 0]

# Function to perform KMeans clustering and segment data
def clustered_segmentation(X_train, X_test, age_column, n_clusters):
    kmeans = KMeans(n_clusters=n_clusters, random_state=seed_num, n_init=30)
    X_train['age_cluster'] = kmeans.fit_predict(X_train[[age_column]])
    X_test['age_cluster'] = kmeans.predict(X_test[[age_column]])
    
    train_clusters = {i: X_train[X_train['age_cluster'] == i] for i in range(n_clusters)}
    test_clusters = {i: X_test[X_test['age_cluster'] == i] for i in range(n_clusters)}
    
    return train_clusters, test_clusters, kmeans

# Function to calculate percentage within tolerance
def calculate_within_tolerance(y_true, y_pred, tolerance):
    within_tolerance = abs(y_true - y_pred) / y_true <= tolerance
    percentage_within_tolerance = within_tolerance.mean() * 100
    return percentage_within_tolerance

# Initialize lists to store metrics across all cluster evaluations
cluster_range = list(range(2, 21))  # 2 to 20 clusters
mse_list = []
mae_list = []
rmse_list = []
mape_list = []
rmspe_list = []
r2_list = []
within_10_list = []
within_50_list = []
within_100_list = []

# Loop over different numbers of clusters
for n_clusters in cluster_range:
    print(f'\n--- Evaluating {n_clusters} Clusters ---')

    # Perform clustering
    train_clusters, test_clusters, kmeans = clustered_segmentation(X_train.copy(), X_test.copy(), 'DRIVER1_AGE', n_clusters)
    
    # Initialize storage for metrics and predictions
    all_y_test = []
    all_y_pred = []

    # Train and evaluate model on each cluster
    for cluster_id in train_clusters.keys():
        train_data = train_clusters[cluster_id].drop('age_cluster', axis=1)
        test_data = test_clusters[cluster_id].drop('age_cluster', axis=1)
        y_train_cluster = y_train.loc[train_clusters[cluster_id].index]
        y_test_cluster = y_test.loc[test_clusters[cluster_id].index]

        if len(train_data) < 2 or len(test_data) < 2:
            print(f"Cluster {cluster_id} too small to evaluate, skipping this cluster.")
            continue

        model = XGBRegressor(random_state=2024)
        model.fit(train_data, y_train_cluster)
        y_pred_cluster = model.predict(test_data)
        
        # Combine predictions and true values for overall evaluation
        all_y_test.extend(y_test_cluster.tolist())
        all_y_pred.extend(y_pred_cluster.tolist())

    # Convert lists to numpy arrays for evaluation
    all_y_test = np.array(all_y_test)
    all_y_pred = np.array(all_y_pred)

    # Calculate overall metrics for the current number of clusters
    overall_mse = mean_squared_error(all_y_test, all_y_pred)
    overall_mae = mean_absolute_error(all_y_test, all_y_pred)
    overall_rmse = np.sqrt(overall_mse)
    overall_mape = np.mean(np.abs((all_y_test - all_y_pred) / all_y_test)) * 100
    overall_rmspe = np.sqrt(np.mean(np.square((all_y_test - all_y_pred) / all_y_test))) * 100
    overall_r2 = r2_score(all_y_test, all_y_pred)
    overall_within_10_percent = calculate_within_tolerance(all_y_test, all_y_pred, 0.10)
    overall_within_50_percent = calculate_within_tolerance(all_y_test, all_y_pred, 0.50)
    overall_within_100_percent = calculate_within_tolerance(all_y_test, all_y_pred, 1.00)

    # Store metrics for plotting
    mse_list.append(overall_mse)
    mae_list.append(overall_mae)
    rmse_list.append(overall_rmse)
    mape_list.append(overall_mape)
    rmspe_list.append(overall_rmspe)
    r2_list.append(overall_r2)
    within_10_list.append(overall_within_10_percent)
    within_50_list.append(overall_within_50_percent)
    within_100_list.append(overall_within_100_percent)

    # Print overall metrics for this dataset
    print(f'\nOverall Metrics for {n_clusters} Clusters:')
    print(f'Mean Squared Error: {overall_mse}')
    print(f'Mean Absolute Error: {overall_mae}')
    print(f'Root Mean Squared Error: {overall_rmse}')
    print(f'Mean Absolute Percentage Error: {overall_mape:.2f}%')
    print(f'Root Mean Squared Percentage Error: {overall_rmspe:.2f}%')
    print(f'R-squared: {overall_r2:.2f}')
    print(f'Percentage within 10%: {overall_within_10_percent:.2f}%')
    print(f'Percentage within 50%: {overall_within_50_percent:.2f}%')
    print(f'Percentage within 100%: {overall_within_100_percent:.2f}%')
```
